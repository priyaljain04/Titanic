```python
import sys
import numpy as np
import pandas as pd
import matplotlib
import scipy as sp
import sklearn

#Common Model Helpers
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn import feature_selection
from sklearn import model_selection
from sklearn import metrics

data_raw = pd.read_csv('train.csv')
data_val = pd.read_csv('test.csv')
print(data_raw.info())
print(data_val.info())

print('train column with null values',data_raw.isnull().sum())
print('test column with null values',data_val.isnull().sum())
data_cleaner = [data_raw,data_val]
for dataset in data_cleaner:
    dataset['Age'].fillna(dataset['Age'].median(),inplace = True)
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0],inplace = True)
    dataset['Fare'].fillna(dataset['Fare'].median(),inplace = True)
    
drop_columns = ['PassengerId','Cabin','Ticket','Name']
data_raw.drop(drop_columns,axis = 1,inplace = True)

print('Train column with null values',data_raw.isnull().sum())
print('Test column with null values',data_val.isnull().sum())


label = LabelEncoder()
for dataset in data_cleaner:
    dataset['Sex_code'] = label.fit_transform(dataset['Sex'])
    dataset['Embarked_code'] = label.fit_transform(dataset['Embarked'])

Target = ['Survived']

data_x = ['Pclass','Age','Sex','SibSp','Parch','Fare']
data_x_calc = ['Pclass','Age','SibSp','Parch','Fare','Sex_code','Embarked_code']
data_raw[data_x_calc]

print('Train column with null values',data_raw.isnull().sum())
print('Test column with null values',data_val.isnull().sum())
```

```python
train_x,test_x,train_y,test_y = model_selection.train_test_split(data_raw[data_x_calc],data_raw[Target],random_state = 0)
```
